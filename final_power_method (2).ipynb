```python
import numpy as np
#75 114 105 115 104 110 97
row = int(input("enter the no. of rows: "))
#row: to get the no. of rows in matrix
column = int(input("enter the no. of column: "))
#column to get the no. of column in matrix

mat=[]
#mat to store the data in matrix form
check=True
for i in range(row):
    mat.append(list(map(int,input("enter the element of row {}: ".format(i+1)).split(" "))))
    #in above we input the entire row at a time
    #conditon the element with spaces will be counted as a different element
    if len(mat[i])!=column:
        print("error")
        break
    #if no. of element is no equal to no. of column that was expected an error will be poped
#above loop is to input the data in matrix

itr = [] #var to store the eigrn vector in iteration
for i in range(row):
    itr.append([1])
#above loop is to create a unitry list i.e. all element are "1"
itr = np.matrix(itr)
#converting the unitry list to matrix
while True:
    count=0
    itr1 = mat*itr 
    mitr=itr1.max()#finding the greatest element of the matrix
    itr1=itr1/mitr #dividing the eigen vector with greatest element of the matrix
    for i in range(row):
        if round(itr1.item(i),4)==round(itr.item(i),4):
            #checking if the value are equal in precision of 4
            count+=1
    if count==row:
        break
        #exiting the loop when repeation is occured in eigen vector
    itr=itr1 
#loop till the repeation in eigen value is found 
#Mprint("Eigen Value: ",'\n',mitr)
#print("Eigen Vector: \n",itr1)

team={}
for i in range(len(itr1)):
    team["team"+str(i+1)]=itr1.item(i)
    itr1.item(i)

rank=dict(sorted(team.items(),key=lambda x:x[1],reverse=True))

j=1
for i in rank:
    print("rank {} =".format(j),i)
    j+=1
   
```
